```python
import pandas as pd
df_monthlyPop = pd.read_csv('./data/monthly_pop.csv',  encoding='cp949', na_values="  ")
df_monthlyPop
```

```python
df_monthlyPop = df_monthlyPop.apply(lambda x: x.str.strip(), axis=1)
```

```python
df_monthlyPop_Total = df_monthlyPop[df_monthlyPop[' 항목 ']=='총인구수']
df_monthlyPop_Total.head()
```

```python
df_monthlyPop_Total =df_monthlyPop_Total.transpose()
df_monthlyPop_Total = df_monthlyPop_Total.drop([df_monthlyPop_Total.index[1],df_monthlyPop_Total.index[2]]) # 항목, 단위 행 삭제
df_monthlyPop_Total = df_monthlyPop_Total.rename(columns=df_monthlyPop_Total.iloc[0])
df_monthlyPop_Total = df_monthlyPop_Total.drop(df_monthlyPop_Total.index[0])
df_monthlyPop_Total.head()
```

```python
df_monthlyPop_Total.reset_index(drop=False, inplace=True)
```

```python
dateSeries = df_monthlyPop_Total['index'].astype('str').str.replace(" ","").str.replace("월","").str.replace(".","")
dateSeries = pd.to_datetime(dateSeries, format = '%Y%m')
df_monthlyPop_Total['date'] = dateSeries
df_monthlyPop_Total.head()
```

```python
df_monthlyPop_Total.shape
```

```python
df_onlyPop = df_monthlyPop_Total.drop(columns=['index','date'],axis =1)
df_onlyPop =df_onlyPop.apply(lambda x: x.str.replace(",",""), axis=1)
df_onlyPop = df_onlyPop.apply(pd.to_numeric)
df_onlyPop['date'] = df_monthlyPop_Total['date']
```

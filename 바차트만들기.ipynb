```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
load_filename = 'data/20190509_mask_POS.csv'
data_pos = pd.read_csv(load_filename)
```

```python
data_pos
```

```python
pie_chart_list=['구군','Sales (Bag)','Sales Amt','유통사','금액']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#66ccff','silver', 'lightgray']
for i in range(len(pie_chart_list)):
    plt.figure(figsize=(15,10))
    data=data_pos.groupby(pie_chart_list[i]).count()['UPC']
    # plt.subplot(2,4,i+1)
    plt.pie(data,colors=colors,wedgeprops = {"edgecolor" : "white",
                      'linewidth': 2,
                      'antialiased': True},
                      
                      autopct='%.2f%%')
plt.show()
```

```python
plt.rc('font', size=20)        # 기본 폰트 크기
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=15)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=20)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=20)  # 범례 폰트 크기
plt.rc('figure', titlesize=10) # figure title 폰트 크기

colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0','#66ccff','silver', 'lightgray']
plt.figure(figsize=(20,10))
data=data_pos.groupby(['지역']).sum()['금액']
# plt.subplot(2,4,i+1)
plt.bar(data.index,height=data,color=colors)
plt.xticks(rotation= 45)
plt.show()
```

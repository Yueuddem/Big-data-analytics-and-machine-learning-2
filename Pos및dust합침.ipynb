```python
df_mask=pd.read_csv('20190509_mask_POS.csv')
df_air_2017=pd.read_csv('SURFACE_air_pollution_2017.csv',encoding='cp949')
df_air_2018=pd.read_csv('SURFACE_air_pollution_2018.csv',encoding='cp949')



df_1702Q_dust=pd.read_csv('1803Q_dust.csv')
```

```python
df_1702Q_dust['지역시']=0
```

```python
#서울 ,인천, 경기도 나누기
for i in range(len(df_1702Q_dust)) :  
  if df_1702Q_dust['지역'][i].split()[0]=='서울':
    df_1702Q_dust['지역시'][i]='서울특별시'
  elif df_1702Q_dust['지역'][i].split()[0]=='인천':
    df_1702Q_dust['지역시'][i]='인천광역시'
  else :
    df_1702Q_dust['지역시'][i]='경기도'

df_1702Q_dust
```

```python
#측정일시 시간 빼고 짜르기
for i in range(len(df_1702Q_dust)) :
    df_1702Q_dust['측정일시'][i]=str(df_1702Q_dust['측정일시'][i])[:4]+'-'+str(df_1702Q_dust['측정일시'][i])[4:6]+'-'+str(df_1702Q_dust['측정일시'][i])[6:8]



df_1702Q_dust
```

```python
df_date_dust=pd.DataFrame(df_1702Q_dust.groupby(['측정일시','지역시']).mean())
df_date_dust=df_date_dust.drop('측정소코드',axis=1)
```

```python
df_date_dust.to_csv('df_date_1803Q.csv')
```
